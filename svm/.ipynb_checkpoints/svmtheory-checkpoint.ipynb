

# ✅ **SVM & Naive Bayes — Simple Jupyter Notebook Answers**

---

## 1. What is a Support Vector Machine (SVM)?

A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression. It finds the best separating line or hyperplane that creates the maximum margin between different classes.

---

## 2. Difference between Hard Margin and Soft Margin SVM

Hard Margin SVM does not allow any misclassification and works only when data is perfectly separable.
Soft Margin SVM allows some misclassifications using the C parameter and works well on noisy real-world data.

---

## 3. Mathematical intuition behind SVM

SVM tries to maximize the margin, which is equal to 2 divided by the norm of the weight vector. It solves an optimization problem to find the best hyperplane that gives the largest margin between classes.

---

## 4. Role of Lagrange Multipliers in SVM

Lagrange Multipliers help convert SVM’s constrained optimization problem into a dual problem. This makes it easier to compute support vectors and apply kernel trick.

---

## 5. What are Support Vectors?

Support Vectors are the data points that lie closest to the decision boundary. These points directly decide the position and orientation of the hyperplane.

---

## 6. What is a Support Vector Classifier (SVC)?

A Support Vector Classifier (SVC) is SVM used for classification problems.

---

## 7. What is a Support Vector Regressor (SVR)?

SVR is the regression version of SVM. It fits the best line within a margin called the epsilon tube.

---

## 8. What is the Kernel Trick?

Kernel Trick is a technique that allows SVM to work in higher-dimensional space without actually computing the transformation. It helps SVM handle non-linear data.

---

## 9. Compare Linear, Polynomial, and RBF Kernel

Linear kernel works for linearly separable data.
Polynomial kernel captures curved boundaries.
RBF kernel handles highly non-linear data and is most commonly used.

---

## 10. Effect of the C Parameter in SVM

Low C creates a larger margin but allows more errors.
High C creates a smaller margin and tries to classify every point correctly.

---

## 11. Role of Gamma in RBF Kernel SVM

Low gamma creates a smooth boundary.
High gamma creates a complex, highly curved boundary.

---

## 12. What is Naïve Bayes? Why is it called “Naive”?

Naive Bayes is a probabilistic classifier based on Bayes’ Theorem.
It is called "naive" because it assumes all features are independent of each other.

---

## 13. What is Bayes’ Theorem?

Bayes’ Theorem calculates the probability of an event based on prior knowledge.
Formula: P(A|B) = [P(B|A) × P(A)] / P(B)

---

## 14. Differences between Gaussian, Multinomial, and Bernoulli Naive Bayes

Gaussian NB is used for continuous numerical features.
Multinomial NB is used for count-based features like word counts.
Bernoulli NB is used for binary features such as word present or not present.

---

## 15. When should you use Gaussian Naive Bayes?

Gaussian NB is used when the features are continuous and follow a normal distribution.

---

## 16. Key assumptions made by Naive Bayes

All features are independent.
Each feature contributes equally to the output.
Conditional probability remains fixed.

---

## 17. Advantages and Disadvantages of Naive Bayes

Advantages: Very fast, works well on text data, needs less training data.
Disadvantages: Independence assumption is often unrealistic, may give poor accuracy if features are correlated.

---

## 18. Why is Naive Bayes good for text classification?

Because text data is high dimensional and mostly independent. It is fast, handles sparse data well, and works well with word counts.

---

## 19. Compare SVM and Naive Bayes for classification tasks

SVM provides high accuracy but is slower and needs more computation.
Naive Bayes is very fast but may give lower accuracy when features are correlated.

---

## 20. How does Laplace Smoothing help in Naive Bayes?

Laplace Smoothing prevents zero probability by adding 1 to word counts. This helps the model avoid undefined probabilities.
